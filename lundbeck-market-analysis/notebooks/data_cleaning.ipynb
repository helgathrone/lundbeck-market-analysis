```python
df_ppp = pd.read_csv("ppp.csv", sep=',')
df_ppp.head()
```

```python
df_ppp.info()
```

```python
ppp_clean = df_ppp[['Reference area','TIME_PERIOD','OBS_VALUE']]

ppp_clean = ppp_clean.rename(columns={
    'Reference area':'country',
    'TIME_PERIOD':'year',
    'OBS_VALUE':'ppp_sales'
})
```

```python
df_all
```

# Lundbeck 2023 and 2024 business performance

Uploading data about business performance in 2023 and 2024 years from Lundbeck annual reports by the extraction of the revenue tables by medicine names from pdf files, cleaning and converting it to .csv with pandas and camelot.

```python
from google.colab import files
uploaded = files.upload()
```

```python
import camelot

tables = camelot.read_pdf(
    "lundbeck_2023.pdf",
    pages="15",
    flavor="stream"
)
```

```python
tables = camelot.read_pdf(
    "lundbeck_2023.pdf",
    pages="15",
    flavor="stream",
    table_areas=["450,600,850,200"]
)

print(len(tables))
```

```python
camelot.plot(tables[0], kind="grid")
```

```python
tables[0].df
```

```python
df = tables[0].df

df.iloc[6,0] = "Vyepti®"
df.iloc[9,0] = "Sabril®"

df = df.drop(index=[0, 1, 2] + list(range(16,25)))
df = df.drop(columns=[2])
df = df.reset_index(drop=True)
df.rename(columns={0: 'TOTAL REVENUE', 1: '2023', 3: '2022', 4: 'Growth', 5: 'Growth (CER)'}, inplace=True)

df
```

```python
df.to_csv("lundbeck_2023_revenue_table.csv", index=False)
```
